##1. Теоретическая справка
**Численное дифференцирование** — совокупность методов приближённого вычисления значения производной некоторой функции, заданной таблично или имеющей сложное аналитическое выражение.

 Численное дифференцирование применяется тогда, когда функцию трудно или невозможно продифференцировать аналитически. Например, необходимость в численном дифференцировании возникает в том случае, когда функция задана не формулой, а таблицей или алгоритмом вычисления в произвольной точке.

Кроме того, формулы численного дифференцирования широко используются при разработке вычислительных методов решения многих
задач, например, при решении дифференциальных уравнений, поиске
решений систем нелинейных уравнений, поиске точек экстремума функций и т.д. Основной подход при построении формул численного дифференцирования — это аппроксимация функции.

 Предположим, что в
окрестности точки $x$ функция $f(x)$ аппроксимируется некоторой другой
функцией $fn(x)$, причем $k$-тая производная $f
(k)$$n(x)$ легко вычисляется.

Естественно в этом случае воспользоваться приближенной формулой:

##*****$f(k)(x) ≈ f_n^k(x)$.*****


---


Для вывода формул численного дифференцирования заменят данную функцию $y=f(x)$ на отрезке $[a,b]$ интерполирующей функцией $P_n(x)$ (полином), а затем полагают $f^I(x)=P_n^I(x)$.
Аналогично и для $f^{II}(x)$.

Если известна погрешность интерполирования $R(x)=f(x)-P_n(x)$, то погрешность численного дифференцирования $r(x)=f^I(x)-P_n^I(x)=R^I(x)$, т.е. погрешность производной интерполирующей функции равна производной от погрешности этой
функции

**Методы численного дифференцирования:**

1. Численное дифференцирование с помощью многочлена Лагранжа.
2. Метод неопределённых коэффициентов.
3. Разложение в ряд Тейлора.

##2. Постановка задачи
Используя выражения для первой и второй разностной центральной производной, определить оптимальный шаг разбиения и вычислить значения, доопределив краевые точки. Промежуток [0,1]

$$y=sin^2(x)/(1+x)$$

В качестве **входных параметров** используются:
* Целевая функция $f(x)$ заданная аналитически

**Результатом** работы является:
* Оптимальный шаг разбиения $h$ при использовании первой и второй центральной разностной производной
* Значения для первой и второй центральной разностной производной с оптимальным шагом
* Графики функции, первых и вторых производных.

**Численная производная из расположения в ряд Тейлора**

Пусть нам данны три точки $x_{i-1},x_i,x_{i+1}$ и значения в них $f(x_{i-1}),f(x_i),f(x_{i+1})$. Шаг между точками пусть равен h.


Тогда из разложения функции $f(x_{i-1})$ в районе точки $x_i$:
$$
f(x_{i-1})=f(x_i)+f'(x_i)*(x_{i-1}-x_i)+f''(Ϛ)\frac{(x_{i-1}-x_i)^2}{2!}
$$
или же
$$
f_{i-1}=f_i+f_i'*(-h)+f''(Ϛ)\frac{h^2}{2!}
$$
отсюда получаем значение **первой производной по левой разности**:
$$
f_i'=\frac{f_i-f_{i-1}}{h}+\frac{h}{2}f''(Ϛ)
$$
где $\frac{h}{2}f``(Ϛ)=O(h^1)$ - погрешность.

**Первая производная по правой разности:**
$$
f_i'=\frac{f_{i+1}-f_{i}}{h}+\frac{h}{2}f''(Ϛ)
$$


### 3. Решение и точность

В качестве входных данных используем:
* Целевая функция $f(x)$ , заданная символьно (аналитически)
* Отрезок $[0,1]$ по которому будем дифферинцировать

**Результатом** работы является:
* Оптимальный шаг разбиения h при использовании первой и второй центральной разностной производной
* Значения для первой и второй центральной разностной производной с оптимальным шагом
* Графики функции, первых и вторых производных.

**Численная производная из расположения в ряд Тейлора**

Пусть нам данны три точки $x_{i-1},x_i,x_{i+1}$ и значения в них $f(x_{i-1}),f(x_i),f(x_{i+1})$. Шаг между точками пусть равен h.


Тогда из разложения функции $f(x_{i-1})$ в районе точки $x_i$:
$$
f(x_{i-1})=f(x_i)+f'(x_i)*(x_{i-1}-x_i)+f''(ξ)\frac{(x_{i-1}-x_i)^2}{2!}
$$
или же
$$
f_{i-1}=f_i+f_i'*(-h)+f''(ξ)\frac{h^2}{2!}
$$
отсюда получаем значение **первой производной по левой разности**:
$$
f_i'=\frac{f_i-f_{i-1}}{h}+\frac{h}{2}f''(ξ)
$$
где $\frac{h}{2}f``(ξ)=O(h^1)$ - погрешность.

**Первая производная по правой разности:**
$$
f_i'=\frac{f_{i+1}-f_{i}}{h}+\frac{h}{2}f''(ξ)
$$

Разложим теперь $f_{i−1}$ и $f_{i+1}$ до $h^3$:
$$
f_{i-1}=f_i-f_i'*h+f_i''\frac{h^2}{2!}-f(ξ)'''\frac{h^3}{3!};x_{i-1}\leq x_i
$$
$$
f_{i+1}=f_i+f_i'*h+f_i''\frac{h^2}{2!}+f(ξ)'''\frac{h^3}{3!};x_{i}\leq x_{i+1}
$$

Если найдем их разность, то получим:
$$
f_i'=\frac{f_{i+1}-f_{i-1}}{2h}+\frac{h^2}{6}(f'''(ξ^+)+f'''(ξ^-))
$$
Это **центральная разность**.

Если сложить разложения $f_{i−1}$ и $f_{i+1}$ и затем выразить $f''(x)$, получим:
$$
f_i''=\frac{f_{i+1}-2f_i+f_{i-1}}{h^2}-h^2\frac{{f^{IV}(ξ^+)+f^{IV}(ξ^-)}}{24}
$$


Погрешность: 
$$
h^2\frac{{f^{IV}(ξ^+)+f^{IV}(ξ^-)}}{24}=O(h^2)
$$

###Задание 1 
Определить оптимальный шаг разбиения h при использовании первой центральной разностной производной, считая, что погрешность определения функции составляет 10% от ее максимального значения на заданном отрезке.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import diff, Symbol, lambdify, exp, sin
from sympy.plotting import plot as sym_plot

#(sin(x))^2 /(1+x) 

fig, ax = plt.subplots(figsize=(10, 6))

def researched_func(x_n):
    return (sin(x))**2 /(1+x) 



x_domain = (0, 1)
x = Symbol('x')
func_sympy = (sin(x))**2 /(1+x) 



def h_from_first_diff(func, domain, point):
    func_lambda = lambdify(x, func)
    max_value = max(func_lambda(domain[0]), func_lambda(domain[1]))

    error_rate = 0.1 * max_value

    # Max number of partitions
    h_max = 20

    df = diff(func, x)
    lam_df = lambdify(x, df)

    df_point = lam_df(point)
    error_h = []

    for i in range(1, h_max):
        h = (domain[1] - domain[0]) / i
        df_teylor = (func_lambda(point + h) - func_lambda(point - h)) / (2 * h)
        if df_teylor + error_rate < df_point:
            df_teylor += error_rate
        elif df_teylor - error_rate > df_point:
            df_teylor -= error_rate
        else:
            return h


def task1():
    h = []
    for i in range(11):
        h.append(h_from_first_diff(func_sympy, x_domain, i / 10))

    avg_h = 0
    for i in h:
        avg_h += i
    avg_h = avg_h / len(h)
    print(f"avg value h = {avg_h}")

    h_x = np.linspace(0, 1, 11)
    ax.plot(h_x, h, color='r')
    ax.grid(True)
    ax.title.set_text("Graph of the dependence of h on the selected point")
    plt.show()
    return h


h1 = task1()
```

**4.2** Определить оптимальный шаг разбиения h при использовании второй центральной разностной производной, считая, что погрешность определения функции составляет $<10$% от ее максимального значения на заданном отрезке.

```python
fig, ax2 = plt.subplots(figsize=(10, 6))

def h_from_second_diff(func, domain, point):
    func_lambda = lambdify(x, func)
    max_value = max(func_lambda(domain[0]), func_lambda(domain[1]))

    error_rate = 0.1 * max_value

      # Max number of partitions
    h_max = 30

    df = diff(func, x)
    d2f = diff(df, x)
    lam_d2f = lambdify(x, d2f)

    df_point = lam_d2f(point)
    error_h = []

    for i in range(1, h_max):
        h = (domain[1] - domain[0]) / i
        df_teylor = (func_lambda(point + h) - 2 * func_lambda(point) + func_lambda(point - h)) / (h ** 2)
        if df_teylor + error_rate < df_point:
            df_teylor += error_rate
        elif df_teylor - error_rate > df_point:
            df_teylor -= error_rate
        else:
            return h


h_from_second_diff(func_sympy, x_domain, 0.1)


def task2():
    h = []
    for i in range(11):
        h.append(h_from_second_diff(func_sympy, x_domain, i / 10))

    avg_h = 0
    for i in h:
        avg_h += i
    avg_h = avg_h / len(h)
    print(f"avg value h = {avg_h}")

    h_x = np.linspace(0, 1, 11)
    ax2.plot(h_x, h, color='r')
    ax2.grid(True)
    ax2.title.set_text("Graph of the dependence of h on the selected point")
    plt.show()
    return h


h2 = task2()
```

**4.3** Используя выражение для первой центральной разностной производной, вычислить ее значения с оптимальным шагом. Доопределить краевые точки значением первой разностной производной той же точности.

```python
def first_diff(func, domain, h):
    
    fig, ax3 = plt.subplots(figsize=(10, 6))
    func_lambda = lambdify(x, func)

    count_of_points = len(h)
    x_arr = np.linspace(domain[0], domain[1], count_of_points)

    y_arr = []

    #numeric value at the left point
    df_teylor_left = ((func_lambda(domain[0] + h[0]) - func_lambda(domain[0])) / (2*h[0]))
    y_arr.append(df_teylor_left)

    for i in range(1, count_of_points-1):
        df_teylor = (func_lambda(x_arr[i] + h[i]) - func_lambda(x_arr[i] - h[i])) / (2*h[i])
        y_arr.append(df_teylor)

    #numeric value at the right point
    df_teylor_right = ((func_lambda(domain[1]) - func_lambda(domain[1] - h[-1])) / (2*h[-1]))
    y_arr.append(df_teylor_right)

    ax3.plot(x_arr, y_arr, color='r')
    ax3.grid(True)
    ax3.title.set_text("First diff")
    plt.show()


first_diff(func_sympy, x_domain,h1)
```

**4.4** Используя выражение для второй центральной разностной производной, вычислить ее значения с оптимальным шагом. 

```python
def second_diff(func, domain, h):
    func_lambda = lambdify(x, func)

    count_of_points = len(h)
    x_arr = np.linspace(domain[0], domain[1], count_of_points)

    y_arr = []

    #numeric value at the left point
    f_left = (func_lambda(domain[0] + h[0]) - 2*func_lambda(domain[0]) + func_lambda(domain[0] - h[0])) / (h[0])**2
    y_arr.append(f_left)

    for i in range(1, count_of_points-1):
        f = (func_lambda(x_arr[i] + h[i]) - 2*func_lambda(x_arr[i]) + func_lambda(x_arr[i] - h[i])) / (h[i])**2
        y_arr.append(f)

    #numeric value at the left point
    f_right = (func_lambda(domain[1] + h[-1]) - 2*func_lambda(domain[1]) + func_lambda(domain[1] - h[-1])) / (h[-1])**2
    y_arr.append(f_right)

    plt.plot(x_arr, y_arr)
    plt.grid(True)
    plt.title("Sec diff")
    plt.show()

second_diff(func_sympy, x_domain, h2)
```
